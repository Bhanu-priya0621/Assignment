Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different
airlines.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('flight_price.xlsx')

# calculate mean and median of flight prices
mean_price = df['Price'].mean()
median_price = df['Price'].median()

print(f"The mean price of flights is {mean_price}.")
print(f"The median price of flights is {median_price}.")

# create histogram to visualize distribution of prices
plt.hist(df['Price'], bins=20)
plt.xlabel('Flight Prices')
plt.ylabel('Frequency')
plt.show()
```




To examine how the prices of flights vary by airline, you can use a boxplot. The boxplot is a useful tool for comparing the distribution of a variable (in this case, price) across different categories (in this case, airlines). Here is an example code that shows how to create a boxplot to compare the prices of different airlines:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('flight_price.xlsx')

# Create a boxplot to compare the prices of different airlines
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['Airline'] == airline]['Price'] for airline in df['Airline'].unique()])
plt.xticks(range(1, len(df['Airline'].unique()) + 1), df['Airline'].unique(), rotation=90)
plt.xlabel('Airline')
plt.ylabel('Price')
plt.title('Flight Prices by Airline')
plt.show()
```
