Q1. What is the mathematical formula for a linear SVM?


The mathematical formula for a linear SVM can be written as:

y(x) = w^T x + b

where:

y(x) is the predicted class label for input x
w is the weight vector that defines the separating hyperplane
x is the input vector
b is the bias term
T is the transpose operator
The SVM algorithm aims to find the optimal values of w and b such that the margin between the positive and negative support vectors is maximized. This can be expressed as the following optimization problem:

minimize (1/2) ||w||^2 subject to y_i(w^T x_i + b) >= 1 for i = 1, ..., n

where:

||w|| is the L2 norm of the weight vector
n is the number of training examples
x_i is the i-th input vector
y_i is the corresponding class label (+1 or -1)

Q2. What is the objective function of a linear SVM?

The objective function of a linear SVM is to find the hyperplane that maximizes the margin between the two classes of data. This can be formulated as a constrained optimization problem, where the objective is to minimize the norm of the weight vector subject to the constraint that all training examples are classified correctly. The objective function can be expressed as:

minimize 1/2 * ||w||^2
subject to yi(w^T xi + b) >= 1 for i = 1, ..., n

where w is the weight vector, b is the bias term, xi is the ith training example, yi is its corresponding label (+1 or -1), and n is the number of training examples. The term ||w||^2 represents the squared norm of the weight vector, which controls the complexity of the model and the size of the margin. The constraint ensures that all training examples are correctly classified and lie on the correct side of the hyperplane.

Q3. What is the kernel trick in SVM?

The kernel trick is a technique used in SVMs to transform the input data into a higher-dimensional space, where it can be more easily separated by a hyperplane. The kernel function calculates the dot product of the transformed feature vectors in the higher-dimensional space, without actually computing the transformation explicitly, which can be computationally expensive or even impossible for very high-dimensional spaces. The kernel trick allows the SVM to operate in the original feature space, while implicitly operating in the higher-dimensional space defined by the kernel function. This technique can improve the accuracy of the SVM, especially when the input data is not linearly separable in the original feature space. Some commonly used kernel functions in SVMs include the linear kernel, polynomial kernel, radial basis function (RBF) kernel, and sigmoid kernel.

Q4. What is the role of support vectors in SVM Explain with example


In SVM, the support vectors are the data points that define the margin between the two classes. These data points are the closest to the decision boundary and play a critical role in determining the hyperplane. The hyperplane is constructed in such a way that it maximizes the margin between the two classes while minimizing the misclassification error.

For example, consider a binary classification problem where we want to classify whether a customer will buy a product or not based on their age and income. We have the following data points:

These three data points lie on the margin and have the most influence on the hyperplane. The other two data points are not support vectors, but they still contribute to the position of the hyperplane. The SVM algorithm finds the best hyperplane that maximizes the margin between the two classes while keeping the misclassification error as low as possible.

Hyperplane:
In SVM, a hyperplane is a decision boundary that separates the data points into different classes. For a binary classification problem, the hyperplane is a line in 2D space, a plane in 3D space, and a hyperplane in higher dimensions.

Marginal plane:
The marginal plane is a parallel plane that is equidistant from the hyperplane and passes through the support vectors. The distance between the marginal plane and the hyperplane is called the margin, which is maximized in SVM.


The hyperplane is the line in the middle, and the marginal planes are the two dotted lines on either side. The margin is the distance between the two marginal planes.

Hard margin:
In hard margin SVM, the data points are linearly separable, and the goal is to find the hyperplane that maximizes the margin between the classes without allowing any misclassifications.

Soft margin:
In soft margin SVM, the data points are not always linearly separable, and the goal is to find the hyperplane that maximizes the margin between the classes while allowing some misclassifications.

In soft margin SVM, some data points are misclassified, but the margin is still maximized. The regularization parameter C controls the trade-off between margin maximization and misclassification penalty. A higher value of C means less tolerance for misclassifications and a narrower margin, while a lower value of C means more tolerance for misclassifications and a wider margin.

Q6. SVM Implementation through Iris dataset.

~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing set
~ Train a linear SVM classifier on the training set and predict the labels for the testing set
~ Compute the accuracy of the model on the testing set
~ Plot the decision boundaries of the trained model using two of the features
~ Try different values of the regularisation parameter C and see how it affects the performance of
the model.

Bonus task: Implement a linear SVM classifier from scratch using Python and compare its
performance with the scikit-learn implementation.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2] # we only take the first two features for visualization purposes
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a linear SVM classifier
svm = LinearSVC(random_state=42)
svm.fit(X_train, y_train)

# Predict the labels for the testing set
y_pred = svm.predict(X_test)

# Compute the accuracy of the model on the testing set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Plot the decision boundaries of the trained model using two of the features
w = svm.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(4, 8)
yy = a * xx - (svm.intercept_[0]) / w[1]

plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.plot(xx, yy, 'k-')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Linear SVM on Iris Dataset')
plt.show()

# Try different values of the regularization parameter C and see how it affects the performance of the model
C_values = [0.1, 1, 10, 100, 1000]
accuracies = []
for C in C_values:
    svm = LinearSVC(C=C, random_state=42)
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    print("C:", C, "Accuracy:", accuracy)

plt.figure(figsize=(10, 6))
plt.plot(C_values, accuracies)
plt.xscale('log')
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.title('Impact of regularization parameter C on accuracy')
plt.show()
```
