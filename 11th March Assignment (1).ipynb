Q2: Differentiate between one-tailed and two-tailed tests.

One-tailed tests allow for the possibility of an effect in one direction. Two-tailed tests test for the possibility of an effect in two directions—positive and negative.



In hypothesis testing, Type 1 and Type 2 errors are the two possible errors that can occur when making a decision about the null hypothesis.

Type 1 error, also known as a false positive, occurs when the null hypothesis is rejected even though it is actually true. This means that the researcher concludes there is a significant effect or difference when in reality there is not. The probability of making a Type 1 error is denoted by alpha (α) and is set by the researcher prior to conducting the hypothesis test.

Type 2 error, also known as a false negative, occurs when the null hypothesis is not rejected even though it is actually false. This means that the researcher fails to detect a significant effect or difference when one actually exists. The probability of making a Type 2 error is denoted by beta (β).

Here are some examples of scenarios for each type of error:

Type 1 error: A company is testing a new drug for its effectiveness in treating a specific illness. The null hypothesis is that the drug has no effect. The researchers conduct a hypothesis test and reject the null hypothesis, concluding that the drug is effective in treating the illness. However, it later turns out that the drug has no actual effect, meaning the researchers made a Type 1 error by rejecting the null hypothesis when it was actually true.

Type 2 error: A researcher is testing a new educational program designed to improve student performance on a standardized test. The null hypothesis is that the program has no effect. The researcher conducts a hypothesis test but fails to reject the null hypothesis, concluding that the program has no effect on student performance. However, it later turns out that the program is actually effective in improving student performance, meaning the researcher made a Type 2 error by failing to reject the null hypothesis when it was actually false.

Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.



Suppose a medical test has a 95% accuracy rate, meaning that if a person has a disease, the test will correctly identify them as having it 95% of the time, and if they don't have the disease, the test will correctly identify them as not having it 95% of the time. The prevalence of the disease in the population is 2%. If a person tests positive for the disease, what is the probability that they actually have it?

To solve this problem using Bayes' theorem, we first need to define the events and probabilities involved:

A = person has the disease
B = person tests positive for the disease
P(A) = 0.02 (prior probability of having the disease)
P(B|A) = 0.95 (probability of testing positive given that the person has the disease)
P(B|not A) = 0.05 (probability of testing positive given that the person does not have the disease)

We want to find P(A|B), the probability of having the disease given that the person tested positive. By Bayes' theorem, we have:

P(A|B) = P(B|A) * P(A) / P(B)

To calculate P(B), the probability of testing positive, we can use the law of total probability:

P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)

where P(not A) = 1 - P(A) = 0.98 (probability of not having the disease).

Substituting in the given values, we get:

P(B) = 0.95 * 0.02 + 0.05 * 0.98 = 0.058

Now we can calculate P(A|B):

P(A|B) = P(B|A) * P(A) / P(B) = 0.95 * 0.02 / 0.058 = 0.3276

Therefore, the probability of a person actually having the disease given that they tested positive is approximately 32.76%.

Margin of error = (z-score) x (standard deviation of the sample / square root of sample size)

Sample size affects the margin of error in two ways. First, as the sample size increases, the standard error of the mean decreases, which means that the margin of error decreases. Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.



Null hypothesis: The true population mean weight loss is equal to or less than 0 pounds.
Alternative hypothesis: The true population mean weight loss is greater than 0 pounds.

The test statistic for a one-sample t-test is given by:

t = (sample mean - hypothesized population mean) / (standard error of the mean)

where the standard error of the mean is equal to the standard deviation of the sample divided by the square root of the sample size.

Substituting in the given values, we get:

t = (6 - 0) / (2.5 / sqrt(50)) = 12

Using a t-distribution table with 49 degrees of freedom (since we have 50 participants and one parameter estimated from the sample), we find that the critical t-value for a one-tailed test at a 95% confidence level is approximately 1.677.

Since our calculated t-value of 12 is greater than the critical t-value of 1.677, we can reject the null hypothesis and conclude that the weight loss drug is significantly effective at a 95% confidence level. This means that the sample mean weight loss of 6 pounds is unlikely to occur by chance if the drug is not effective, and it provides evidence in support of the alternative hypothesis that the true population mean weight loss is greater than 0 pounds.

reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to
determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.


Null hypothesis (H0): The average reaction time is not significantly different between participants who consume caffeine and those who do not consume caffeine.
Alternative hypothesis (Ha): The average reaction time is significantly different between participants who consume caffeine and those who do not consume caffeine.

We can use a two-sample t-test to compare the means of two independent samples. Since we only have one sample, we can assume that the population standard deviation is unknown and use the t-distribution.

The formula for the t-test statistic is:

t = (x̄ - μ) / (s / √n)

Where:
x̄ = sample mean = 0.25 seconds
μ = hypothesized population mean = 0 (assuming no effect of caffeine)
s = sample standard deviation = 0.05 seconds
n = sample size = 30

Using a t-table or calculator, the critical t-value for a two-tailed test at a 90% confidence level with 28 degrees of freedom is 1.701.

Calculating the t-test statistic:

t = (0.25 - 0) / (0.05 / √30)
t = 7.745

Since the calculated t-value (7.745) is greater than the critical t-value (1.701), we can reject the null hypothesis and conclude that there is a significant difference in reaction time between participants who consume caffeine and those who do not consume caffeine.

Interpretation: The study provides strong evidence to suggest that caffeine has a significant effect on reaction time among the participants in the study
